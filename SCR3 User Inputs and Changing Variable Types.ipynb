```python
# user Input:
name = input("Input your name here: ")
# User input with type conversion
age = int(input("Input your age here: "))

# Output with print() function
print("Hello, \nmy name is", name, "and \nmy age is", age)
```

```python
# Convert string to float
no_years = float(7)

# Mathematical operations can only be done with numeric types (no strings)
age_future = age + no_years

# Output with print() function
print("Hello, \nmy name is", name, "and \nmy age is", age, "\n and in", no_years, "years, I will be", age_future)
```
