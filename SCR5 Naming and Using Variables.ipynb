```python
# When Assigning variables, the variable being assigned must be on the left.
x = 7
y = 10
print(x)
```

```python
# The variable being assigned cannot be on the LHS of the equal sign
7 = x
print(x)
```

```python
# The variable being assigned can also be on the rhs of the = , but it must already be defined 
x = x + 1
print("Output 1:", x)
x = x + 1
print("Output 2:", x)
w = w + 1
```
