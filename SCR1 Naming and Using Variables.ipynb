```python
# To return information to the user, use the print() function
print(x)
```

```python
# When a print statement is used, by default it creates a new line, for example:
print(x)
print(x+x)
```

```python
# If you do not create a new line whenever you print you can enter an alternate way to end your print statement
# by defining an 'end' parameter in the print function
# For example:
print(x, end = ' ')
print(x + x)
```

```python
# To combine strings with variables in a print statement, 
# separate the string from the variable using a comma (this creates a space)
name = "Darla"
age = 42
print("Hello, my name is", name,"and I am", age, "years old.")
```

```python
# now lets say you want to create a new line partway through a print statment without creating
# a second print statement. For this you will use what is known as a newline character, \n within your string variable
name = "Darla"
age = 42
print("Hello, \nmy name is", name,"and \nI am", age, "years old.")
```
