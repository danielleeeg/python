```python
x = 2
y = 3
# Addition: +
a = x + y
print(a)
```

```python
# Subtraction: -
b = x - y
print(b)
```

```python
# Multiplication: *
c = x * y
print(c)
```

```python
# Division: /
d = x / y
print(d)
```

```python
# Exponent: **
f = x ** y
print(f)
```

```python
# Modulus: %
g = y % x
print(g)
```

```python
# Flooring://
h = y // x
print(h)
```

```python
# Scientific notation: e
#  1 x 10 ^ 5
i = 1e5
print(i)
```

```python
# Order of operations: PEMDAS
# 1. Parentheses: ()
# 2. Unary minus: 3**-y (negation, not subraction)
# 3. Multiplication, division, modulus: *, /, %
# 4. Addition, subtraction: +, -
# 5. Equvalent operations are evaluated left to right: 2 * 4 / 7 % 9
j = 8*(2 + 3)/7 + (4*-6)
print(j)
```
